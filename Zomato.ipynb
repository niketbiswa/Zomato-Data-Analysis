Zomato Data Analysis Using Python

Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Creating Data Frame

```python
dataframe = pd.read_csv("Zomato data .csv")
dataframe.head()
```

Converting Rate Column into Float and removing denominator

```python
def handleRate(value):
    value=str(value).split('/')
    value=value[0]
    return float(value)

dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

Summary of Data Frame

```python
dataframe.info()
```

Types Of Resturant

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("Type of Resturant")
```

Ratings

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes':grouped_data})
plt.plot(result,c="green",marker="o")
plt.xlabel("Type of Resturant",c="red",size="20")
plt.ylabel("Votes",c="red",size="20")
```

Distribution Of Ratings

```python
plt.hist(dataframe['rate'],bins=5)
plt.title("Ratings Distribution")
plt.show
```

Approx Cost

```python
couple_data = dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

Online Order

```python
plt.figure(figsize=(6,6))
sns.boxplot(x = 'online_order', y = 'rate' , data = dataframe,
hue='online_order',legend=False   )
```

Offline orders received lower ratings in compariosn to online orders,which obtained excellent ratings.

Conclusion

```python
pivot_table = dataframe.pivot_table(
    index='listed_in(type)', 
    columns='online_order', 
    fill_value=0,  
    aggfunc='size'  
)

plt.figure(figsize=(10, 8))
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='.1f')  

# Adding labels and title
plt.title("Heatmap of Ratings by Online Order and Type")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```

Dining Resturant primaliry accepts offline order,whereas cafes primaliry accepts online orders.This suggests that clients prefer to place order in person at resturants,but prefer online ordering at cafes.
